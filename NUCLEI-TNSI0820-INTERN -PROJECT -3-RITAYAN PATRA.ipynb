```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        print("Train Result:")
        print(f"accuracy score: {accuracy_score(y_train, pred):.4f}\n")
        print(f"Classification Report: Precision: {precision_score(y_train, pred)}\nRecall Score: {recall_score(y_train, pred)}\nF1 score: {f1_score(y_train, pred)}\n")
        print(f"Confusion Matrix:  {confusion_matrix(y_train, clf.predict(X_train))}\n")
        
    elif train==False:
        pred = clf.predict(X_test)
        print("Test Result:")        
        print(f"accuracy score: {accuracy_score(y_test, pred)}\n")
        print(f"Classification Report:  Precision: {precision_score(y_test, pred)}\nRecall Score: {recall_score(y_test, pred)}\nF1 score: {f1_score(y_test, pred)}\n")
        print(f"Confusion Matrix:  {confusion_matrix(y_test, pred)}\n")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
print('RANDOM FOREST CLASSIFIER')
print_score(rf, X_train, y_train, X_test, y_test, train=True)
print("-------------------------------------------------------------")
print_score(rf, X_train, y_train, X_test, y_test, train=False)
```

```python
dt= DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
print("DECISION TREE CLASSIFIER")
print_score(dt, X_train, y_train, X_test, y_test, train=True)
print("---------------------------------------------------------------------")
print_score(dt, X_train, y_train, X_test, y_test, train=False)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
lg=LogisticRegression()
lg.fit(X_train,y_train)
y_predict=lg.predict(X_test)
print("LOGISTIC REGRESSION CLASSIFIER")
print_score(lg, X_train, y_train, X_test, y_test, train=True)
print("---------------------------------------------------------------------")
print_score(lg, X_train, y_train, X_test, y_test, train=False)
```

```python
svc = SVC()
svc.fit(X_train, y_train)
print('SUPPORT VECTOR CLASSIFIER')
print_score(svc, X_train, y_train, X_test, y_test, train=True)
print("---------------------------------------------------------------------")
print_score(svc, X_train, y_train, X_test, y_test, train=False)
```

```python
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train, y_train)
print('KNN CLASSIFIER')
print_score(knn, X_train, y_train, X_test, y_test, train=True)
print("---------------------------------------------------------------------")
print_score(knn, X_train, y_train, X_test, y_test, train=False)
```

```python
gauss = GaussianNB()
gauss.fit(X_train, y_train)
print('NAIVE BYTES CLASSIFIER')
print_score(gauss, X_train, y_train, X_test, y_test, train=True)
print("---------------------------------------------------------------------")
print_score(gauss, X_train, y_train, X_test, y_test, train=False)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
print('XGB CLASSIFIER')
print_score(xgb, X_train, y_train, X_test, y_test, train=True)
print("---------------------------------------------------------------------")
print_score(xgb, X_train, y_train, X_test, y_test, train=False)
```

```python
#Using ANN

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(X_train.shape[1], activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
r = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, class_weight={0:1, 1:2})
```

```python
training_score = model.evaluate(X_train, y_train)
testing_score = model.evaluate(X_test, y_test)

print(f"TRAINING SCORE: {training_score}")
print(f"TESTING SCORE: {testing_score}")
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(2, 2, 1)
plt.plot(r.history['loss'], label='Loss',linewidth=4)
plt.plot(r.history['val_loss'], label='val_Loss',linewidth=4)
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(r.history['accuracy'], label='accuracy',linewidth=4)
plt.plot(r.history['val_accuracy'], label='val_accuracy',linewidth=4)
plt.legend()
```

```python
#Using ANN

def accuracy_score_details(label, prediction, train=True):
    if train:
        print("Train Result:")
        print(f"Accuracy Score: {accuracy_score(label, prediction) * 100:.2f}%")
        print("Classification Report:", end='')
        print(f"\nPrecision Score: {precision_score(label, prediction) * 100:.2f}%")
        print(f"Recall Score: {recall_score(label, prediction) * 100:.2f}%")
        print(f"F1 score: {f1_score(label, prediction) * 100:.2f}%")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, prediction)}\n")
        
    elif train==False:
        print("Test Result:")        
        print(f"Accuracy Score: {accuracy_score(label, prediction) * 100:.2f}%")
        print("Classification Report:", end='')
        print(f"\nPrecision Score: {precision_score(label, prediction) * 100:.2f}%")
        print(f"Recall Score: {recall_score(label, prediction) * 100:.2f}%")
        print(f"F1 score: {f1_score(label, prediction) * 100:.2f}%")
        print(f"Confusion Matrix: \n {confusion_matrix(label, prediction)}\n")
```

```python
y_train_predict = model.predict(X_train)
y_test_predict = model.predict(X_test)

accuracy_score_details(y_train, y_train_predict.round(), train=True)
accuracy_score_details(y_test, y_test_predict.round(), train=False)
```
